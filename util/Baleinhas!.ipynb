```python
especies = ['bottlenose_dolphin','beluga','humpack_whale','blue_whale','false_killer_whale','dusky_dolphin']
```

```python
# Função que receberá uma lista de especies e filtrará fotos e dataSet.
def filtragemBaleinhas (list) :
    # Caminho relativo total(com nome do arquivo) de entrada.
    caminhoDataSetEntrada = input('Digite o caminho total do dataSet(.csv) de entrada : ')
    # Carregamento do DataSet
    df = pd.read_csv(caminhoDataSetEntrada)
    # Caminho relativo (sem nome dos arquivos) das imagens para mover.
    caminhoImagens = input('Digite onde estão as imagens para filtrar : ')
    caminhoImagens = caminhoImagens + '\\'
    # Nome dataSet de saída já filtrado.
    nome = input('Digite o nome do Arquivo do dataSet de saída :  ')
    nome = nome + '.csv'
    # Destino dos arquivos de fotos e csv.
    destino = input('Digite o caminho de destino para os arquivos fotos e csv : ')
    destino = destino + '\\'
    data = pd.DataFrame()
    for i in especies :
        print(f'\n Espécies Filtradas: ',i)
        data = data.append(df.loc[df.species == str(i)])
    
    for j in data.index : 
        source = caminhoImagens + df.loc[j][0]
        destination = destino + df.loc[j][0]
        
        if os.path.isfile(source):
            shutil.move(source, destination)
    return data.to_csv(destino+nome,index=False)
```

```python
filtragemBaleinhas(especies)
```
